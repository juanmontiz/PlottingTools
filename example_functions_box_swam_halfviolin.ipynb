## Plot **any number** of arrays with corresponding labels:

### Some useful tools and functions for making different types of plots (box plots, violin plots, etc.).

- The box (violin) and swarm **palettes can be different**! In case they are the same, the box (violin) palette is set to be more translucid than the swarm so that the points can be clearly seen.
- Also, a custom palette can also be provided.
- The arrays can have different numbers of elements.
- The code has to be executed twice for the **font_scale** to take effect (the ghosts in the machine...).
- If saveplot=True, the plot is saved in PNG, PDF, and SVG formats. Provide the filename **without extension**.

### Some of the available predefined palettes:
* Categorical:
    - "tab10"
    - "deep" (default palette)
    - "muted"
    - "pastel"
    - "dark"
    - "colorblind"

* Sequential:
    - "Blues"
    - "Greens"
    - "Oranges"
    - "Purples"
    - "Reds"
    - "Greys"

* Diverging:
    - "coolwarm"
    - "RdBu"
    - "Spectral"
    - "PuOr"
    - "BrBG"

* Matplotlib:
    - "viridis"
    - "plasma"
    - "cividis"
    - "magma"
    - "inferno"

```python
# I create different arrays to plot
NPOINTS = 50

# Create arrays with random data
np.random.seed(42)
ARR_1 = 0.002*np.random.rand(NPOINTS - 10) + 0.004
ARR_1[-1] = np.max(ARR_1) * 1.1 # Outlier! 
ARR_2 = 0.0025*np.random.rand(NPOINTS - 30) + 0.0045 # Arrays can have different number of elements!!
ARR_3 = 0.004*np.random.rand(NPOINTS) + 0.005
ARR_3[-1] = np.min(ARR_3) * 0.8 # Outlier!
```

```python
# Create the dataset from arrays and assign labels
dataset = [ARR_1, ARR_2, ARR_3]
labels = ['label 1', 'label 2', 'label 3']
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
palette='tab10'
plot_half_violin_box_swarm(ax=ax,
                            dataset=dataset,
                            labels=labels,
                            y_min=None, y_max=None,
                            h_line=None,
                            box_palette=palette,
                            swarm_palette=palette,
                            violin_palette=palette,
                            bias=0.1,
                            font_scale=1.1,
                            point_size=5,
                            plot_title='Title?',
                            y_axis_label='Y-axis label?',
                            show_p_values=True,
                            ygrid=True,
                            saveplot=False,
                            filename='filename',
                            dpi=300)
plt.tight_layout()
plt.show()
```
