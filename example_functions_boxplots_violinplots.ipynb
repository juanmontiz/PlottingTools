## Plot **any number** of arrays with corresponding labels:

### Some of the available palettes:
* Categorical:
    - "deep" (default palette)
    - "muted"
    - "pastel"
    - "dark"
    - "colorblind"

* Sequential:
    - "Blues"
    - "Greens"
    - "Oranges"
    - "Purples"
    - "Reds"
    - "Greys"

* Diverging:
    - "coolwarm"
    - "RdBu"
    - "Spectral"
    - "PuOr"
    - "BrBG"

* Matplotlib:
    - "viridis"
    - "plasma"
    - "cividis"
    - "magma"
    - "inferno"

### Also...
- The box (violin) and swarm **palettes can be different**! In case they are the same, the box (violin) palette is set to be more translucid than the swarm in order to clearly see the points.
- Also, a custom palette can also be provided.
- The arrays can have different number of elements.
- The code has to be execute twice for the **font_scale** to take effect (the ghosts in the machine...).
- If saveplot=True the the plot is saved in png, pdf, and svg formats. Provide the filename **without extension**.

# Box plots

```python
# I create different arrays to plot
NPOINTS = 50

# Create arrays with random data
ARR_1 = 0.002*np.random.rand(NPOINTS - 10) + 0.004
ARR_1[-1] = np.max(ARR_1) * 1.3 # Outlier! 
ARR_2 = 0.0025*np.random.rand(NPOINTS - 30) + 0.004 # Arrays can have different number of elements!!
ARR_3 = 0.004*np.random.rand(NPOINTS) + 0.006
ARR_4 = 0.006*np.random.rand(NPOINTS) + 0.007
ARR_4[-1] = np.min(ARR_4) * 0.4 # Outlier!
```

```python
# Create the dataset from arrays and assign labels
datasets = [ARR_1, ARR_2, ARR_3, ARR_4]
labels = ['label 1', 'label 2', 'label 3', 'label 4']
```

### Example showing horizontal lines (xgrid=True), no $p$-values (show_p_values=False), not definig the y-axis legend nor title, and no plot saving.

```python
fig, ax = plt.subplots(figsize=(10, 7))
plot_boxes_generalized(ax, datasets, labels,
                           box_palette="deep", swarmplot_palette="deep",
                           font_scale=1.4,
                           point_size=6,
                           show_p_values=False,
                           xgrid=True,
                           saveplot=False)
```

### Example showing horizontal lines (xgrid=True), no $p$-values (show_p_values=False), and no plot saving.

```python
fig, ax = plt.subplots(figsize=(10, 7))
plot_boxes_generalized(ax, datasets, labels,
                           box_palette="deep", swarmplot_palette="deep",
                           font_scale=1.4,
                           point_size=6,
                           plot_title='Here is the title of the plot',
                           y_axis_label='Mean velocity [m/s]',
                           show_p_values=True,
                           xgrid=False,
                           saveplot=True,
                           filename='the_best_boxplot',
                           dpi=450)
```
